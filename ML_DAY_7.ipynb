```python
import matplotlib.pyplot as plt
```

```python
# multiple points
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# two points
xpoints = [0,6]
ypoints = [0,250]
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# default x - points will be increments of 1 [0,1,2,3,4...]
# xpoints = [0,6]
ypoints = [10, 20, 30, 40, 50]
plt.plot(ypoints)
plt.show()
```

```python
# format strings fmt: marker|line|color
# mfc is marker face color

ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, marker = 'v', ms = 15, color = 'g', mfc = 'b') #ms - marker size
plt.show()
```

```python
""" '-' solid line
':' dotted line
'--' dashed line
'-.' dash-dot line"""

ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, 'o-.k', linewidth='10')
plt.show()
```
