```python
import matplotlib.pyplot as plt
```

```python
# multiple points
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# two points
xpoints = [0,6]
ypoints = [0,250]
plt.plot(xpoints, ypoints)
plt.show()
```

```python
# default x - points will be increments of 1 [0,1,2,3,4...]
# xpoints = [0,6]
ypoints = [10, 20, 30, 40, 50]
plt.plot(ypoints)
plt.show()
```

```python
# format strings fmt: marker|line|color
# mfc is marker face color

ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, marker = 'v', ms = 15, color = 'g', mfc = 'b') #ms - marker size
plt.show()
```

```python
""" '-' solid line
':' dotted line
'--' dashed line
'-.' dash-dot line"""

ypoints = [10,20,30,40,50,60]
plt.plot(ypoints, 'o-.k', linewidth='2')
plt.show()
```

```python
# multiple points
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]

x2points = [1,2,3,4,5]
y2points = [17,27,38,67,89]

# plt.plot(xpoints, ypoints)
# plt.plot(x2points, y2points)
plt.plot(xpoints, ypoints, 'o-.k', x2points, y2points, '*--b')
plt.show()
```

```python
# Title & Labels
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]
plt.plot(xpoints, ypoints, 'v-.r')
plt.title("Sales per day")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.show()
```

```python
# formatting

xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]
plt.plot(xpoints , ypoints)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("sale per day", fontdict = font1)
plt.xlabel("day" , fontdict= font2 )
plt.ylabel("sale", fontdict= font2)
plt.title("sale per day")
plt.show()
```

```python
xpoints = [1,2,3,4,5]
ypoints = [10,20,30,40,50]
plt.plot(xpoints , ypoints)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("sale per day", fontdict = font1, loc = 'right')
plt.xlabel("day" , fontdict= font2 )
plt.ylabel("sale", fontdict= font2)

plt.show()
```

```python
import datetime


current_date= datetime.date.today()
xpoints=[current_date]
for i in range(1, 5):
    next_date = current_date + datetime.timedelta(days=i)
    print(next_date)
    xpoints.append(next_date)

#xpoints = [1,2,3,4,5]
ypoints = [10,20,30,90,50]
plt.plot(xpoints,ypoints)

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sales Per Day", fontdict = font1)
plt.xlabel('Day',fontdict=font2)
plt.ylabel('Sale',fontdict=font2)
plt.title('Sales Per Day')
plt.show()
```

```python
# subplots 1st arg is row , 2nd arg is col , 3rd arg is index of the current plot

x1 = [1,2,3,4,5]
y1= [1,2,3,4,5]
#plot 1
plt.subplot(1,2,1)
plt.plot(x1,y1)

#plot 2
x2 = [1,2,3,4,5]
y2 = [17,27,38,67,89]
plt.subplot(1,2,2)
plt.plot(x2,y2)

plt.show()
```
